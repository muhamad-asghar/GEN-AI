```python
import pandas as pd
df=pd.read_csv('Amazon Sale Report.csv')
```

```python
import spacy


# Load the SpaCy language model
nlp = spacy.load("en_core_web_sm")

# Read the CSV file into a DataFrame


# Display the first few rows of the DataFrame to inspect column names
print(df.head())

# Display the column names
print(df.columns)

# Define the tokenization function
def tokenize_text(text):
    # Process the text with SpaCy
    doc = nlp(text)
    # Extract word tokens
    word_tokens = [token.text for token in doc]
    # Extract sentence tokens
    sentence_tokens = [sent.text for sent in doc.sents]
    return word_tokens, sentence_tokens

# Replace 'text_column_name' with the actual column name that contains the text data
df['word_tokens'], df['sentence_tokens'] = zip(*df['Fulfilment'].apply(tokenize_text))

# Display the DataFrame with the new tokenized columns
print(df.head())
```

# Removing StopWords

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
def remove_stopwords(text):
    new_text = []
    for word in text.split():
        if word in stopwords.words('english'):
            new_text.append('')
        else:
            new_text.append(word)
    x = new_text[:]
    new_text.clear()
    return " ".join(x)
```

```python
remove_stopwords(df['Fulfilment'][6])
```

```python
import nltk
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
nltk.download('punkt')
# Load NLTK's PorterStemmer
stemmer = PorterStemmer()

# Load the CSV file into a DataFrame


# Define a function to apply stemming to each text entry
def stem_text(text):
    # Tokenize the text
    words = word_tokenize(text)
    # Apply stemming to each word
    stemmed_words = [stemmer.stem(word) for word in words]
    # Join the stemmed words back into a single string
    stemmed_text = ' '.join(stemmed_words)
    return stemmed_text

# Apply stemming to the entire corpus
df['stemmed_text'] = df['Fulfilment'].apply(stem_text)

# Optionally, you can save the stemmed corpus to a new CSV file
df.to_csv('stemmed_corpus.csv', index=False)

```

# Lemmatization

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer

# Download NLTK resources

nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')

# Initialize the lemmatizer
lemmatizer = WordNetLemmatizer()

# Function to perform lemmatization
def lemmatize_text(text):
    # Tokenize the text
    words = word_tokenize(text)
    # Get the part of speech (POS) tag for each token
    pos_tags = nltk.pos_tag(words)
    # Map POS tags to WordNet POS tags
    def get_wordnet_pos(tag):
        if tag.startswith('J'):
            return wordnet.ADJ
        elif tag.startswith('V'):
            return wordnet.VERB
        elif tag.startswith('N'):
            return wordnet.NOUN
        elif tag.startswith('R'):
            return wordnet.ADV
        else:
            return wordnet.NOUN  # Default to noun if POS tag is not found
    # Lemmatize each token with its POS tag
    lemmatized_words = [lemmatizer.lemmatize(word, get_wordnet_pos(pos_tag)) for (word, pos_tag) in pos_tags]
    # Join the lemmatized words back into a single string
    lemmatized_text = ' '.join(lemmatized_words)
    return lemmatized_text

# Read the CSV file into a DataFrame


# Apply lemmatization to the entire corpus
df['lemmatized_text'] = df['Fulfilment'].apply(lemmatize_text)

# Save the DataFrame with the lemmatized corpus to a new CSV file
df.to_csv('lemmatized_corpus.csv', index=False)
```
